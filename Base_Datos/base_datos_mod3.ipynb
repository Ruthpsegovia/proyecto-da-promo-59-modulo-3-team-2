```python
# IMPORTAR LIBRERIAS
import os #trabajar con el sistema operativo y variables de entorno
from dotenv import load_dotenv #cargar variable de entorno

import pandas as pd ## para convertir los datos en formato tabla
import mysql.connector ## para conectar Paython con MySQL
import numpy as np ## para convertir formato nan de Paython a none de MySQL
from mysql.connector import Error ## captura errores de MySQL
```

```python
load_dotenv() #carga las variables del entorno .env; devuelve un true o false


#MySQL
MYSQL_HOST = os.getenv("MYSQL_HOST")
MYSQL_USER = os.getenv("MYSQL_USER")
MYSQL_PASSWORD = os.getenv("MYSQL_PASSWORD")
```

```python
# CONEXIÓN Y CREACIÓN DE BASE DE DATOS 
try:
    cnx = mysql.connector.connect(
        host= '127.0.0.1' ,
        user= 'root',
        password= '111729',
    )
    print('Conexión exitosa')
except Error as tipo_error:
    print('Error al conectar:', tipo_error)

try:
    mycursor = cnx.cursor()
    query = "CREATE DATABASE IF NOT EXISTS ABC_Corporation_db"
    mycursor.execute(query)
    print("Base de datos creada con éxito")
except:
    print("Error.")
```

```python
# CREACIÓN DE TABLAS EN BASE DE DATOS
mycursor.execute("USE ABC_Corporation_db") # Seleccionamos la base de datos

# 1-Tabla Departments
try:
    query_Departments = '''
    CREATE TABLE IF NOT EXISTS Departments (
        ID_Department INT AUTO_INCREMENT,
        Department VARCHAR(50),
        PRIMARY KEY (ID_Department)
    );
    '''
    mycursor.execute(query_Departments)
    cnx.commit()  # Guardar cambios
    print("Tabla Departments creada correctamente")
except Exception as e:
    print("Error al crear tabla Departments:", e)
    cnx.rollback()

# 2-Tabla JobRoles
try:
    query_JobRoles = '''
    CREATE TABLE IF NOT EXISTS JobRoles (
        ID_JobRole INT AUTO_INCREMENT,
        JobRole VARCHAR(50),
        PRIMARY KEY (ID_JobRole)
    );
    '''
    mycursor.execute(query_JobRoles)
    cnx.commit()
    print("Tabla JobRoles creada correctamente")
except Exception as e:
    print("Error al crear tabla JobRoles:", e)
    cnx.rollback()

# 3-Tabla Employees
try:
    query_Employees = '''
    CREATE TABLE IF NOT EXISTS Employees (
        EmployeeNumber INT,
        Age FLOAT,
        AgeGroup VARCHAR(20),
        Gender VARCHAR(20),
        MaritalStatus VARCHAR(20),
        DistanceFromHome INT,
        Education INT,
        EducationField VARCHAR(50),
        JobLevel INT,
        Attrition VARCHAR(10),
        Attrition_flag INT,
        ID_Department INT,
        JobRole VARCHAR(50),
        PRIMARY KEY (EmployeeNumber),
        FOREIGN KEY (ID_Department) REFERENCES Departments(ID_Department)
    );
    '''
    mycursor.execute(query_Employees)
    cnx.commit()
    print("Tabla Employees creada correctamente")
except Exception as e:
    print("Error al crear tabla Employees:", e)
    cnx.rollback()

# 4-Tabla Salaries
try:
    query_Salaries = '''
    CREATE TABLE IF NOT EXISTS Salaries (
        EmployeeNumber INT,
        HourlyRate INT,
        DailyRate INT,
        MonthlyRate INT,
        MonthlyIncome FLOAT,
        PercentSalaryHike INT,
        StockOptionLevel INT,
        PRIMARY KEY (EmployeeNumber),
        FOREIGN KEY (EmployeeNumber) REFERENCES Employees(EmployeeNumber)
    );
    '''
    mycursor.execute(query_Salaries)
    cnx.commit()
    print("Tabla Salaries creada correctamente")
except Exception as e:
    print("Error al crear tabla Salaries:", e)
    cnx.rollback()

# 5-Tabla Satisfaction
try:
    query_Satisfaction = '''
    CREATE TABLE IF NOT EXISTS Satisfaction (
        EmployeeNumber INT,
        EnvironmentSatisfaction INT,
        JobSatisfaction INT,
        RelationshipSatisfaction INT,
        WorkLifeBalance INT,
        PRIMARY KEY (EmployeeNumber),
        FOREIGN KEY (EmployeeNumber) REFERENCES Employees(EmployeeNumber)
    );
    '''
    mycursor.execute(query_Satisfaction)
    cnx.commit()
    print("Tabla Satisfaction creada correctamente")
except Exception as e:
    print("Error al crear tabla Satisfaction:", e)
    cnx.rollback()

# 6-Tabla WorkConditions
try:
    query_WorkConditions = '''
    CREATE TABLE IF NOT EXISTS WorkConditions (
        EmployeeNumber INT,
        BusinessTravel VARCHAR(30),
        OverTime VARCHAR(10),
        JobInvolvement INT,
        PerformanceRating INT,
        PRIMARY KEY (EmployeeNumber),
        FOREIGN KEY (EmployeeNumber) REFERENCES Employees(EmployeeNumber)
    );
    '''
    mycursor.execute(query_WorkConditions)
    cnx.commit()
    print("Tabla WorkConditions creada correctamente")
except Exception as e:
    print("Error al crear tabla WorkConditions:", e)
    cnx.rollback()

# 7-Tabla Experience
try:
    query_Experience = '''
    CREATE TABLE IF NOT EXISTS Experience (
        EmployeeNumber INT,
        TotalWorkingYears INT,
        NumCompaniesWorked INT,
        YearsAtCompany INT,
        YearsInCurrentRole INT,
        YearsSinceLastPromotion INT,
        YearsWithCurrManager FLOAT,
        TrainingTimesLastYear FLOAT,
        PRIMARY KEY (EmployeeNumber),
        FOREIGN KEY (EmployeeNumber) REFERENCES Employees(EmployeeNumber)
    );
    '''
    mycursor.execute(query_Experience)
    cnx.commit()
    print("Tabla Experience creada correctamente")
except Exception as e:
    print("Error al crear tabla Experience:", e)
    cnx.rollback()
```
